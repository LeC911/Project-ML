```python
def count_unique(stroka):
    unique_values = set(stroka)
    unique_dict = {}
    for i in unique_values:
        unique_dict[i] = stroka.count(i)
    return unique_dict
```

```python
count_unique('aabcddeghttllyyyz')
```

```python
def count_unique(s):
    d = {}
    for i in s:
        if i not in d:
            d[i] = 0
        d[i] += 1
    return d
```

```python
count_unique('aabcddeghttllyyyz')
```

```python
import collections
def count_unique(s):
    return collections.Counter(s)
```

```python
count_unique('aabcddeghttllyyyz')
```

```python
import csv

def read_csv(file):
    with open(file, encoding='utf-8') as r_f:
        reader = csv.reader(r_f)  # считываем csv файл
        lst = list(reader)  # переводим в список
        # zip создает кортеж -> map применяет функцию len к каждой колонке -> max находит максимальную длину
        max_len = (max(map(len, col)) for col in zip(*lst))
        # {:<} каждый следующий (next) элемент выравнивается по левому краю
        # join разделяет пробелами каждый элемент в max_len -> f-строка выравнивает его по правому краю {:>}
        s = f"{{:<{next(max_len)}}} {' '.join(f'{{:>{ml}}}' for ml in max_len)}"
        for i in lst:
            print(s.format(*i))
```

```python
read_csv('titanic.csv')
```

```python
read_csv('Киновзгляды.csv')
```

```python
import pandas as pd
def read_csv(file):
    df = pd.read_csv(file)
    return df.columns.values.tolist() , df.values.tolist()
```
